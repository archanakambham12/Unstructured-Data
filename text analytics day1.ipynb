## Conclusion
    - stemmer helps to reduce the number of distinct words in corpus
    - lemmetizer helps to retain the meaning of the words effectively used when the contect of the word is important
    - stemmer is used when i have an intent of getting lesser words which inturn reduces the total number of features

```python
#combine all the steps and create a udf
#takes one review as input and returns a cleaned review as a string
def clean_txt(sent):
    review_tokens=word_tokenize(sent.lower())
    stop_nltk=stopwords.words("english")
    stop_updated=stop_nltk+["...","..","!!"]+["phone","mobile","lenovo","k8","note","amazon","n't"]
    reviews_updated1=[term for term in review_tokens\
                  if term not in stop_updated\
                  and term not in list(punctuation) and len(term)>2]
    ls=[stemmer_s.stem(word) for word in reviews_updated1]
    res=" ".join(ls)
    return res
```




##  4. trem frequency
- occurance of a word in a doc/ total no. of distinct words present in that doc
- TF(good,0)=2/5 # in document 0, good  appears 2 times out of 5 distinct words
- TF(plain,0)=1/5

## 5. Document Frequency-
- n0.of documents containing the word/ Total no. of Documents
- DF(good)=1/2
- DF(plain)= 1/2
- DF(intresting)=2/2

## compute TFIDF OF GOOD
- TF(good,0)=2/5
-  DF(good)=1/2
- IDF(good)=ln(1/DF)=ln(2)=0.693
- TFIDF(good)=TF*IDF= 2/5* 0.693=0.277259



## inverse document freq- its the logarthmic inverse of DF
- no. of total documents=100
- the is present in 100 documents
- sachin is present in 10 documents
- cover drive is present 20 times

- idf(the)=log(100/100)=0
- idf(sachin)= log(100/10)=1
- idf(cover drive)= 1og(100/20)=0.69

## TF
- 100 documents 
- suppose in one if the documents with 1000 words:
- the 500 times
- sachin 200 times
- tf(the)= 500/1000=0.5
- tf(sachin)= 200/1000=0.2

# tfidf= tf* IDF
- tfidf(the)=0
- tfidf(sachin)=0.2
- tfidf(cover drive)=0.01
